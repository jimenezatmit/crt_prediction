# 1: Feature Extraction and Reduction Tuning 

In this section, the text features (mentions, hashtags, Tweets/Retweets, followees, bio, followee_bios, and follower_bios) will undergo tuning for two hyperparameters used only on the text features: 

1. **n_components** in TruncatedSVD for dimensionality reduction: desired dimensionality of output data
2. **min_df** in TF-IDF vectorizer for feature extraction: terms that have a document frequency strictly lower than min_df will be ignored when building the vocabulary.

The Ridge model with cross-validation is used for tuning. 

```python
# run this cell to define text features to tune 
text_features = ['text', 'domains', 'followees', 'mentions', 'hashtags', 'bio', 'follower_bios', 'followee_bios']

general.set_background('#fff9e9')
```

<font color='blue'>**Variable Descriptions**</font>

**1.2 Tune min_df:** For each text feature, tune min_df. 

*A similar tuning algorithm is used in this section; however, only the start range value for min_df is defined. The end range value for min_df is automatically calculated by the algorithm based on the feature and the best n_components value calculated in Section 1.1*.

4. **ITERS_MINDF** (integer): number of iterations to perform Ridge + cross validation, default is 10
5. **TOP_MINDF** (integer): number of top min_df to print after tuning, default is 10

<font color='blue'>**Folder Descriptions**</font>
1. **RESULTS_FOLDER_HISTORY_MINDF** (string): name of folder to save tuning results

# 3: Feature Selection

In this section, we perform feature selection using ElasticNet and cross-validation. ElasticNet is chosen because    it addresses over-regularization by balancing the penalties of LASSO (*alpha* = 1) and Ridge (*alpha* = 0). 

The feature selection process runs as follows, using the Embedded Method:
1. Each feature is passed through ElasticNet with cross-validation, penalizing the coefficients in the model.
2. Any coefficients that shrink to zero are dropped from the model, indicating that these certain predictors are not helpful in estimating the target. If there are no zero-valued coefficients, the smallest non-zero coefficient is dropped.
3. Continue eliminating predictors until each feature only contains one predictor. 
4. The accuracy of each new set of predictors is calculated using the Ridge model. 
5. Choose the set of "best" predictors for each feature, which is the set of predictors that yield the highest Pearson r correlation coefficient using Ridge.

*Source*: [Medium](https://medium.com/analytics-vidhya/feature-selection-for-dimensionality-reduction-embedded-method-e05c74014aa)

<font color='blue'>**Variable Descriptions**</font>
1. **STOP_FS** (integer): feature selection stops once the selected number of features equals this value, default is 1 
2. **DISPLAY_FS** (boolean): display progress and results if True, default is True
3. **ITERS_FS** (integer): number of iterations to perform Ridge + cross validation after each round of feature selection, default is 10
4. **PARMAS_FS** (dictionary): dictionary containing parameters for ElasticNet ([*ElasticNet specs*](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.ElasticNetCV.html))

<font color='blue'>**Folder Descriptions**</font>
1. **RESULTS_FOLDER_FS** (string): name of folder to save feature selection results 

```python
# TODO: set parameters

STOP_FS = 1 
DISPLAY_FS = True 
ITERS_FS = 10
PARAMS_FS = {
            'l1_ratio': np.linspace(0.00001, 1, 20), # default is np.linspace(0.00001, 1, 20)
            'fit_intercept': True, # default is True
            'alphas': np.logspace(-1, 2, 20), # default is np.logspace(-1, 2, 20)
            'cv': 10, # default is 10
            'max_iter': 100000, # default is 100000
            'random_state': 17, # default is 17
            'n_jobs': -1, # default is -1
            'tol': .00001, # default is .00001
            'verbose': 0 # default is 0
        }

# TODO: set folder paths
RESULTS_FOLDER_FS = ROOT_DIR + RESULTS_FOLDER_NAME + 'selection_new_elastic/'

general.set_background('#efe1e1')
```

```python
# run this cell to perform feature selection
feat_dict = selection.split_columns(master_data_train, MY_TARGET)
selection_results = pd.DataFrame(columns=['feature', 'target_feats', 'ridge_r_value', 'ridge_p_value', 'history'])

for feat, featlist in feat_dict.items():
    
    if feat != 'domains':
        continue
        
    # do not perform feature selection on CRT score
    if feat.startswith("CRT"):
        continue

    results = selection.feature_selection(master_data_train, master_data_test, feat, featlist, MY_TARGET, PARAMS_FS, 
                                          stop=STOP_FS, display=DISPLAY_FS, iters=ITERS_FS)
    selection_results = selection_results.append(results, ignore_index=True)

selection_df, plot, chart = selection.plot(selection_results, RESULTS_FOLDER_FS, MY_TARGET, DATAFRAME_TYPE, DISPLAY_FS)

general.set_background('#fff9e9')
```

# 5: Combine Features

In this section, we combine the best features from feature selection to generate the best possible model using one of three models: 

1. Ridge model
2. LASSO model
3. Random Forests Regressor model

The following protocol is used to combine features to generate the best model:
1. Begin with the feature that yields the highest r value during feature selection
2. Run feature through model pipeline 
3. Add feature that yields the subsequent highest r value during selection and re-run model pipeline
4. Repeat until all features have been added from highest to lowest r values

<font color='blue'>**Variable Descriptions**</font>
1. **MODEL_CF** (string): name of model 
    1. Use "ridge" for Ridge model
    2. Use "lasso" for LASSO model

```python
import glob
import pandas as pd
import json
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import datetime
```

```python
outlet_ratings=pd.read_csv('60 outlets ratings by factcheckers.csv')
outlet_ratings.columns=['domain','score','type','twitter_handle']
outlet_ratings['score']=(outlet_ratings['score']-1)/4.
Domains=set(outlet_ratings['domain'])
```

```python
outlet_ratings['score'].min(),outlet_ratings['score'].max()
```

```python
with open('data/twitter_data/tweeted_domains.csv','w') as outfile:
    outfile.write('screen_name,domain,date,RT,site\n')
    for i,filename in enumerate(glob.glob('/Users/mm1049/Dropbox/Personal/Research/crt-estimator/data/data_merged_20211114/combined_till_20211112_processed/*.txt')):
            try:
                data = json.load(open(filename))
                screen_name = filename.split('/')[-1].replace(".txt","")
                if 'RTs' not in data.keys():
                    data['RTs']=[0]*len(data['sites'])
                for site,date,RT in zip(data['sites'],data['dates'],data['RTs']):
                    domain=''
                    for domain_ in Domains:
                        if site.find(domain_)!=-1:
                            domain=domain_
                    outfile.write('{},{},{},{},{}\n'.format(screen_name,domain,date,RT,site.replace(',','')))


            except Exception as e:
                print (e)
                print ('screen_name',screen_name)
```

```python
tweeted_domains=pd.read_csv('data/twitter_data/tweeted_domains.csv')
```

```python
tweeted_domains
```

```python
df=tweeted_domains
df=pd.merge(left=df,right=outlet_ratings,on='domain',how='left')
```

```python
df_=df.groupby('screen_name').sum().reset_index()
df_=df_[['screen_name','score']]
df_.columns=['screen_name','outlet_score']
df_.to_csv('data/twitter_data/outlet_scores_sum.csv',index=False)
```

```python
df_=df.groupby('screen_name').mean().reset_index()
df_=df_[['screen_name','score']]
df_.columns=['screen_name','outlet_score']
df_.to_csv('data/twitter_data/outlet_scores_mean.csv',index=False)
```
